```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'coulmns')
merged
```

```python
# prompt: Using dataframe dummies: how to get data in binary?

# Convert boolean columns to numeric (0 and 1)
dummies_numeric = dummies.astype(int)

# Convert the numeric DataFrame to a NumPy array
dummies_array = dummies_numeric.to_numpy()

# Convert the NumPy array to binary representation
dummies_binary = dummies_array.tobytes()
dummies_binary

# Now, dummies_binary contains the data in binary format.
# You can further process or save it to a file as needed.

# Example: Saving to a file
# with open("dummies_binary.bin", "wb") as f:
#    f.write(dummies_binary)
```
